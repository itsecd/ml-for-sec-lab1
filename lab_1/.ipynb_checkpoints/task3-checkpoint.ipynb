### Задача №3

На основании данных о клиентах Банка, необходимо построить модель машинного обучения для прогнозирования факта просрочки обязательных платежей по текущему кредиту. Выполняется в Jupyter Notebook. В качестве меры оценки качества модели использовать F1-score. В работе продемонстрировать основные шаги – работа с пропусками, выбросами, типами данных, графический анализ, статистики, формирование выборок, моделирование, тюнинг модели, получение скоров, анализ полученных результатов, интерпретация результата на языке бизнеса.

* Home_Ownership - домовладение
* Annua_Income - годовой доход
* Years_in_current_job - количество лет на текущем месте работы
* Tax_Liens - налоговые обременения
* Number_of_Open_Accounts - количество открытых счетов
* Years_of_Credit_History - количество лет кредитной истории
* Maximum_Open_Credit - наибольший открытый кредит
* Number_of_Credit_Problems - количество проблем с кредитом
* Months_since_last_delinquent - количество месяцев с последней просрочки платежа
* Bankruptcies - банкротства
* Purpose - цель кредита
* Term - срок кредита
* Current_Loan_Amount - текущая сумма кредита
* Current_Credit_Balance - текущий кредитный баланс
* Monthly_Debt - ежемесячный долг
* Credit_Score - кредитный скор
* Credit_Default - факт невыполнения кредитных обязательств (0 - погашен вовремя, 1 - просрочка)

Проверяемые навыки: классическое обучение с учителем, умение интерпретировать используемые методы моделирования, понимание полученного результата.

Данные: test.csv – тестовая выборка, train.csv – обучающая выборка.


```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, recall_score, precision_score, roc_auc_score, make_scorer

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
train.shape, test.shape
```

```python
train.head(3)
```

```python
train.drop('Id', axis=1, inplace=True)
```

```python
sns.boxplot(x=train["Current Loan Amount"])
```

Действительно, в выборке существуют выбросы, избавимся от них.

```python
train = train.loc[train['Current Loan Amount'] != 99999999.0]
```

```python
sns.boxplot(x=train["Current Loan Amount"])
```

Теперь все в норме, продолжаем...

Посмотрим на пропущенные значения в выборке

```python
plt.figure(figsize=(8, 5))
sns.heatmap(train.isnull())
```

```python
for col in train.columns:
    pct_missing = np.mean(train[col].isna())
    if pct_missing:
        print(f'{col} - {round(pct_missing * 100)}%')
```

```python
train['Months since last delinquent'] = train['Months since last delinquent'].fillna(0)
train['Bankruptcies'] = train['Bankruptcies'].fillna(0)
```

```python
train['Annual Income'] = train['Annual Income'].fillna(train['Annual Income'].mean())
train['Credit Score'] = train['Credit Score'].fillna(train['Credit Score'].mean())
```

Для того, чтобы обработать пропуски в колонке years in current job, необходимо избавиться от текста, оставим только числа

```python
train['Years in current job'] = train['Years in current job'].str.extract(r'(\d+)')
train['Years in current job'] = train['Years in current job'].fillna(0)
train['Years in current job'] = train['Years in current job'].astype(int)
```

```python
train.isnull().sum()
```

Пропущенных значений более нет, продолжаем работу

```python
fig, ax = plt.subplots(figsize=(5,5))
sns.countplot(x=train['Credit Default'], palette='Set2')
plt.show()
```

В графике распределения таргета дизбаланс, людей, которые выполнили кредитные обязательства больше, нежели тех, которые не выполнили, на это стоит обратить внимание

```python
train.head(3)
```

У нас есть переменные, которые нуждаются в кадировании, а именно HomeOwnerShip, Purpose & Term

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
train['Home Ownership'] = le.fit_transform(train['Home Ownership'])
train['Purpose'] = le.fit_transform(train['Purpose'])
train['Term'] = le.fit_transform(train['Term'])
```

```python
sns.set(font_scale=1)
corr_with_target = train.corr().iloc[:-1, -1].sort_values(ascending=False)
plt.figure(figsize=(9, 9))
sns.barplot(x=corr_with_target.values, y=corr_with_target.index)
plt.title('Корреляция с таргетом')
```

В целом, предобработка данных закончена, осталось провести такие же манипуляции с test данными и перейдем к подбору модели

```python
plt.figure(figsize=(8, 5))
sns.heatmap(test.isnull())
```

```python
test['Months since last delinquent'] = test['Months since last delinquent'].fillna(0)
test['Bankruptcies'] = test['Bankruptcies'].fillna(0)
test['Annual Income'] = test['Annual Income'].fillna(test['Annual Income'].mean())
test['Credit Score'] = test['Credit Score'].fillna(test['Credit Score'].mean())
test['Years in current job'] = test['Years in current job'].str.extract(r'(\d+)')
test['Years in current job'] = test['Years in current job'].fillna(0)
test['Years in current job'] = test['Years in current job'].astype(int)
test['Home Ownership'] = le.fit_transform(test['Home Ownership'])
test['Purpose'] = le.fit_transform(test['Purpose'])
test['Term'] = le.fit_transform(test['Term'])
```

```python
X, y = train.drop('Credit Default', axis=1), train['Credit Default']
```

Возьмем реализацию градиентного бустинга из CatBoost для предсказания, посмотрим в начале на модель с стандартными параметрами, а затем попробуем ее затюнить для улучшения результата предсказания

```python
X_train, X_test, y_train, y_test = train_test_split(X,  y, stratify=y, test_size=0.2, random_state=42, shuffle=True)
```

```python
model = CatBoostClassifier(verbose=0)
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

```python
print('Train F1: ', f1_score(y_train, model.predict(X_train)))
print('Train Recall:', recall_score(y_train, model.predict(X_train)))
print('Train Precision:', precision_score(y_train, model.predict(X_train)))
print('-' * 35)
print('Test F1: ', f1_score(y_test, y_predict))
print('Test Recall:', recall_score(y_test, y_predict))
print('Test Precision:', precision_score(y_test, y_predict))
```

Модель с точностью 0.72 правильно прогнозирует, выполнит ли клиент обязательства по кредиту или нет. И полнота правильных предсказаний, что уйдет клиент или нет равна ~0.36. Среднегармоническое между Precision и Recall дает ~0.48, что крайне мало, данный результат можно и нужно улучшать. Будем тюнить CatBoost, дабы улулчшить предсказания модели.

Для тюнинга модели будем использовать байесовские оптимизации над гиперпараметрами из библиотеки scikit-optimize.

```python
from sklearn.model_selection import StratifiedKFold
from skopt import BayesSearchCV
from skopt.callbacks import DeadlineStopper, VerboseCallback, DeltaXStopper
from skopt.space import Real, Categorical, Integer
```

```python
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=0)
roc_auc = make_scorer(roc_auc_score, greater_is_better=True, needs_threshold=True)
```

```python
search_spaces = {
    'iterations': Integer(10, 1000),
    'depth': Integer(1, 8),
    'learning_rate': Real(0.01, 1.0, 'log-uniform'),
    'random_strength': Real(1e-9, 10, 'log-uniform'),
    'bagging_temperature': Real(0.0, 1.0),
    'border_count': Integer(1, 255),
    'l2_leaf_reg': Integer(2, 30),
    'scale_pos_weight':Real(0.01, 1.0, 'uniform')
}
```

```python
opt = BayesSearchCV(
    model,
    search_spaces,
    scoring=roc_auc,
    cv=skf,
    n_iter=150,
    n_jobs=-1,
    return_train_score=False,
    refit=True,
    optimizer_kwargs={'base_estimator': 'GP'},
    random_state=42
)
```

```python
opt.fit(X, y)
```

```python
opt.best_params_
```

```python
params = {
    'bagging_temperature': 0.0,
    'border_count': 236,
    'depth': 2,
    'iterations': 610,
    'l2_leaf_reg': 11,
    'learning_rate': 0.08201343297653359,
    'random_strength': 8.465655242677597e-05,
    'scale_pos_weight': 0.8370345431843947,
    'iterations': 1000
}
```

```python
model_tuned = CatBoostClassifier(**params, verbose=0)
model_tuned.fit(X_train, y_train)
y_predict_tuned = model_tuned.predict(X_test)
```

```python
print('Train tuned F1: ', f1_score(y_train, model_tuned.predict(X_train)))
print('Train tuned Recall:', recall_score(y_train, model_tuned.predict(X_train)))
print('Train tuned Precision:', precision_score(y_train, model_tuned.predict(X_train)))
print('-' * 35)
print('Test tuned F1: ', f1_score(y_test, y_predict_tuned))
print('Test tuned Recall:', recall_score(y_test, y_predict_tuned))
print('Test tuned Precision:', precision_score(y_test, y_predict_tuned))
```

Исходя из полученных результатов, байесовские оптимизации над гиперпараметры ухудшили результаты модели, нежели модель, которая была с стандартными параметрами. Можно было бы попробовать другие модели машинного обучения, например, реализацию градиентного бустинга из LightGBM, может стоило попробовать случайные деревья и пробовать тюнить их, так же для сравнения можно было бы реализовать полносвязную нейронную сеть, а смогла ли бы она улучшить результаты, которые были получены? На данном этапе нетюненная модель градиентного бустинга из CatBoost наилучший вариант для предсказания, сделаем предсказания и сохраним их в файл submissions.csv

```python
predictions = model.predict(test)
final = pd.DataFrame(data=test['Id'], columns=['Id'])
final['target'] = predictions
final.to_csv('submission.csv', index=False)
```
